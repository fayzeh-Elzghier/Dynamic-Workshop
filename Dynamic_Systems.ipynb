```python
# ğŸ“˜ Dynamic Systems Exploration
# Interactive Visualization for First and Second Order Systems
# Prepared for Faiza 

# =======================
# ğŸ§© Import Libraries
# =======================
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl
from ipywidgets import interact

# =======================
# âš™ï¸ First-Order System
# =======================
# G(s) = K / (Ï„s + 1)
# Ï„ (tau): Time constant â€” larger Ï„ = slower response
# K: System gain â€” higher K = higher steady-state value

def first_order(K=1.0, tau=1.0):
    sys = ctrl.tf([K], [tau, 1])
    t = np.linspace(0, 10, 500)
    t_out, y_out = ctrl.step_response(sys, t)
    plt.figure(figsize=(7,4))
    plt.plot(t_out, y_out, label=f"K={K}, Ï„={tau}")
    plt.title("ğŸ“ˆ First-Order Step Response")
    plt.xlabel("Time (s)")
    plt.ylabel("Output y(t)")
    plt.grid(True)
    plt.legend()
    plt.show()

print("âœ¨ Move the sliders to see how K and Ï„ affect the first-order response:")
interact(first_order, K=(0.5, 3.0, 0.5), tau=(0.2, 3.0, 0.2));

# =======================
#  Second-Order System
# =======================
# G(s) = Ï‰â‚™Â² / (sÂ² + 2Î¶Ï‰â‚™s + Ï‰â‚™Â²)
# Î¶ (zeta): Damping ratio â€” controls oscillation
# Ï‰â‚™ (omega_n): Natural frequency â€” controls speed of response

def second_order(zeta=0.5, omega_n=2.0):
    sys = ctrl.tf([omega_n**2], [1, 2*zeta*omega_n, omega_n**2])
    t = np.linspace(0, 10, 500)
    t_out, y_out = ctrl.step_response(sys, t)
    plt.figure(figsize=(7,4))
    plt.plot(t_out, y_out, label=f"Î¶={zeta}, Ï‰â‚™={omega_n}")
    plt.title("ğŸ“ˆ Second-Order Step Response")
    plt.xlabel("Time (s)")
    plt.ylabel("Output y(t)")
    plt.grid(True)
    plt.legend()
    plt.show()

print("\nğŸ’« Move the sliders to explore how Î¶ and Ï‰â‚™ affect the second-order response:")
interact(second_order, zeta=(0.0, 2.0, 0.1), omega_n=(0.5, 5.0, 0.5));

# =======================
# ğŸ’¡ Observations (for students)
# =======================
# 1ï¸âƒ£ Increasing Ï„ makes the first-order response slower.
# 2ï¸âƒ£ Higher K raises the steady-state value.
# 3ï¸âƒ£ In second-order systems:
#     - Î¶ < 1 â†’ oscillations (underdamped)
#     - Î¶ = 1 â†’ fastest non-oscillatory response (critical)
#     - Î¶ > 1 â†’ slow non-oscillatory (overdamped)
# 4ï¸âƒ£ Higher Ï‰â‚™ â†’ faster system response.

print("\nğŸ“š Observations:")
print("- Ï„ â†‘ â†’ response slower.")
print("- Î¶ controls damping (oscillations).")
print("- Ï‰â‚™ controls speed.")
print("- K controls output level.")
print("\nâœ… Try different combinations and note how the curves behave!")
```
